```python
df2 = pd.read_csv("C:/python/datas/아파트 평당가.csv")
```

```python
df2.info()
```

```python
df2['전용면적(㎡)'].astype('float')
```

```python
df2['거래금액(만원)'] = df2['거래금액(만원)'].str.strip().replace("'", '')
```

```python
df2['거래금액(만원)'] = df2['거래금액(만원)'].str.replace(",", '').astype('f')
```

```python
df2.info()
```

```python
# df2.insert(3, '전용면적 당 거래금액(㎡)',df2['거래금액(만원)']/df2['전용면적(㎡)'])
df2['전용면적 당 거래금액(㎡)'] = round((df2['거래금액(만원)']) / (df2['전용면적(㎡)']),2)
```

```python
df2
```

```python
df2.시군구 = df2.시군구.str.split().str[1]
df2
```

```python
df3 = round(df2.groupby('시군구').mean(), 2)
df3
```

```python
df3.info()
```

```python
df3['거래금액(만원)'] = df3['거래금액(만원)'].astype('float64')
df3
```

```python
df3.info()
```

```python
df3 = round(df3, 2)
df3
```

```python
seoul_geojson = "https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json"

m = folium.Map(location = [37.57, 126.99], zoom_start = 11)
folium.Choropleth(geo_data = seoul_geojson,
                  data=df3[["시군구", "전용면적 당 거래금액(㎡)"]],
                  columns=["시군구", "전용면적 당 거래금액(㎡)"],
                  key_on="feature.properties.name",
                  fill_color = "PuRd",
                  fill_opacity = 0.4,
                  line_opacity = 1
).add_to(m)

m
```
