```python
# 데이터 읽기
crime_raw_data = pd.read_csv(
    "C:/python/datas/crime_in_Seoul.csv", thousands=',',
)
# null 데이터 삭제
crime_raw_data = crime_raw_data[crime_raw_data['죄종'].notnull()]
# 데이터 확인
crime_raw_data
```

```python
crime_station = crime_raw_data.pivot_table(
    crime_raw_data, 
    index=['구분'], 
    columns=['죄종', '발생검거'], 
    aggfunc=[np.sum]
)
crime_station.head()
```

```python
crime_station.columns = crime_station.columns.droplevel([0, 1])
crime_station.head()
```
