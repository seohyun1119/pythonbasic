```python
# mlp.rc(‘rcParams’, 변경할 값을 키워드 인수로 전달)는 rcParams를 변경해주는 메서드
```

```python
# dict 타입의 변수를 함수(혹은 메서드)의 키워드 인수로 사용하려면 **을 붙여서 전달

global_font = {'family' : 'NanumGothic', 'size' : 12, 'weight' : 'bold'}

# 폰트설정
mpl.rc('font', **global_font) # pass in font dict as kwargs
# mpl.rc('font', 'family' : 'NanumGothic', 'size' : 24, 'weight' : 'bold')와 같이 동작
```

```python
import pandas as pd

# RcParams 중 font

data = {
    "설명" : ["필기체", "폰트", "판타지", "고정폭", "산세리프", "세리프",
           "폰트 크기", "장평", "표시 형태", "대소문자 표기법", "굵기"]
}

df = pd.DataFrame(data, index = ["cursive", "family", "fantasy", "monospace", 
                                "sans-serif", "serif", "size", "stretch",
                                "style", "variant", "weight"])
df.index.name = "attributes"
df
```

```python
plt.title("한글 제목")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("엑스축 레이블")
plt.ylabel("와이축 레이블")
plt.show()
```

```python
# 객체마다 별도의 폰트를 적용
# 폰트 family, color, size: dict 타입으로 정의
# 이 dict 변수를 플롯 메서드의 fontdict 키워드 인수로 전달

title_font = {'family' : 'NanumMyeongjo', 
              'size' : 24, 
              'color' : 'black'}

xlabel_font = {'family' : 'NanumBarunpen', 
              'size' : 18,
              'weight' : 'bold',
              'color' : 'darkred'}

ylabel_font = {'family' : 'NanumBarunGothic', 
              'size' : 12,
              'weight' : 'light',
              'color' : 'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title("한글 제목", fontdict = title_font)
plt.xlabel("엑스 축", fontdict = xlabel_font)
plt.ylabel("와이 축", fontdict = ylabel_font)
plt.show()
```

```python
# Unicode minus 표기법 설정
# 음수 부호를 유니코드와 ASCII 중 어떤 문자로 표기할지 설정
# rc() 메서드로 RcParams 중 axes.unicode_minus=False로 변경하면 ASCII의 문자로 표현

# 유니코드에서 음수 부호 설정

mpl.rc('axes', unicode_minus=False)
```

```python
# 스타일 문자열은 색상(color), 마커(marker), 선 종류(line style)의 순서로 지정
```

```python
data_color = {
    "약자" : ['k', 'w', 'r', 'b', 'g', 'c', 'm', 'y']
}

style_color = pd.DataFrame(data_color, index = ["black", "white", "red", "blue",
                                                "green", "cyan", "magenta", "yellow"])
style_color.index.name = "색 이름"
style_color
```

```python
data_marker = {
    "설명" : ['point marker', 'pixel marker', 'circle marker', 'triangle_down marker',
           'triangle_up marker', 'triangle_left marker', 'triangle_right marker',
           'hline, vline', 'star marker', 'pentagon marker', 'hexagon1 marker', 
           'hexagon2 marker', 'plus marker', 'x marker', 'diamond marker', 
           'thin_diamond marker']
}

style_marker = pd.DataFrame(data_marker, index = ['.', ',', 'o', 'v, 1', '^, 2', '<, 3',
                                                 '>, 4', '_, |', '*', 'p', 'h', 'H', '+',
                                                 'x', 'D', 'd'])
style_marker.index.name = "구분"
style_marker
```

```python
data_linestyle = {
    "설명" : ['solid line', 'dashed line', 'dash-dotted line',
           'dotted line', 'draw nothing']
}

style_linestyle = pd.DataFrame(data_linestyle, index = ["'_' or 'solid'",
                                                       "'--' or 'dashed'",
                                                       "'-.' or 'dashdot'",
                                                       "':' or 'dotted'",
                                                       "'None', ' ', ''"])
style_linestyle.index.name = "구분"
style_linestyle
```

```python
data_etc = {
    "약자" : ['c', 'lw', 'ls', '', 'ms', 'mec', 'mew', 'mfc'],
    "description" : ['선 색상', '선 굵기', '선 스타일', '마커 종류',
                    '마커 크기', '마커 선 색깔', '마커 선 굵기',
                    '마커 내부 색상']
}

style_etc = pd.DataFrame(data_etc, index = ['color', 'linewidth',
                                                 'linestyle', 'marker',
                                                 'markersize', 'markeredgecolor',
                                                 'markeredgewidth', 'markerfacecolor'])
style_etc.index.name = "구분"
style_etc
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c = 'b',
         lw = 5, ls = '--', marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')
plt.title("스타일 적용 예")
plt.show()
```
