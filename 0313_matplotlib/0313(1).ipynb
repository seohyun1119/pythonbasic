```python
# 객체마다 별도의 폰트를 적용
# 폰트 family, color, size: dict 타입으로 정의
# 이 dict 변수를 플롯 메서드의 fontdict 키워드 인수로 전달

title_font = {'family' : 'NanumMyeongjo', 
              'size' : 24, 
              'color' : 'black'}

xlabel_font = {'family' : 'NanumBarunpen', 
              'size' : 18,
              'weight' : 'bold',
              'color' : 'darkred'}

ylabel_font = {'family' : 'NanumBarunGothic', 
              'size' : 12,
              'weight' : 'light',
              'color' : 'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title("한글 제목", fontdict = title_font)
plt.xlabel("엑스 축", fontdict = xlabel_font)
plt.ylabel("와이 축", fontdict = ylabel_font)
plt.show()
```

```python
# 축의 범위 설정하기
# xlim(최소값, 최대값)과 ylim(최소값, 최대값)을 사용
# x축, y축의 최소값과 최대값을 지정

plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
# xticks()이나 yticks(): 틱 설정
# 플롯이나 차트에서 축상의 위치 표시 지점: 틱(tick)
# 틱에 써진 숫자 혹은 글자: 틱 라벨(tick label)

import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x, c)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
# 틱 라벨 문자열에는 $$ 사이에 수학 문자식(Mathematical expressions) 넣을 수 있음

x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x, c)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
# 여러 개의 선 그리기

# x데이터, y데이터, 스타일 문자열을 반복하여 인수로 넘긴다.
# 이 경우에는 하나의 선을 그릴 때처럼 x 데이터나 스타일 문자열을 생략할 수 없음

t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t **3, 'g^-')
plt.show()
```

```python
# 여러 개의 선 그리기
# 하나의 plot()이 아니라 복수의 plot()을 하나의 그림에 겹쳐서 그리기

plt.title("복수의 plot 명령을 한 그림에서 표현")
style1 = {"c" : "b", "lw" : 5, "ls" : "--", "marker" : "o", "ms" : 15,
          "mec" : "g", "mew" : 5, "mfc" : "r"}
style2 = {"c" : "k", "lw" : 3, "ls" : ":", "marker" : "s", "ms" : 10,
          "mec" : "m", "mew" : 5, "mfc" : "c"}

plt.plot([1, 4, 9, 16], **style1)
plt.plot([9, 16, 4, 1], **style2)

plt.show()
```
