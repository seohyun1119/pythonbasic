```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Figure

# Matplotlib는 data를 Figure(e.g. windows, Jupyter widgets, etc.)에 x-y 좌표의 그래프로 표시함
# 각각의 Figure는 하나 이상의 Axes를 가짐
# pyplot.subplots(): Figure, Axes 객체 만들기
# Axes.plot(): Axes에 데이터를 그림

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);
```

```python
# Figure: 전체 그림(그래프)을 의미하는 객체

fig = plt.figure()                   # Axes 없이 Figure만 생성
fig, ax = plt.subplots()             # 1개의 Axes 객체를 갖는 Figure
fig, axs = plt.subplots(2, 2)        # grid 형태의 shape을 갖는 Axes 객체 생성
```

```python
# 그림의 크기는 figsize 인수로 설정

import numpy as np

np.random.seed(0)
f1 = plt.figure(figsize = (10, 2))
plt.title("figure size: (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

```python
# subplot(): 그리드(grid) 형태의 Axes 객체들을 생성
# Figure가 행렬(matrix)이고 Axes가 행렬의 원소

ax1 = plt.subplot(2, 1, 1)
# 여기에서 윗부분에 그릴 플롯 명령 실행

ax2 = plt.subplot(2, 1, 2)
# 여기에서 아랫부분에 그릴 플롯 명령 실행
```

```python
# 위/아래 두 개의 플롯을 하나의 Figure 안에 그리기
# 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()     # 플롯 간의 간격을 자동으로 맞춰준다.
plt.show()
```

```python
# 2x2 형태의 네 개의 플롯
# Axes의 순서는 콘솔 출력 순서와 동일(좌에서 우로, 위에서 아래로)

np.random.seed(0)

plt.subplot(221)              # (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시 가능
plt.plot(np.random.rand(5))
plt.title('axes 1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes 2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes 3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes 4')

plt.tight_layout()
plt.show()
```

```python
# 위 코드를 다르게 표현

fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title('axes 1')
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title('axes 2')
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title('axes 3')
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title('axes 4')

plt.tight_layout()
plt.show()
```

```python
# Axes와 Axis는 다른 객체이므로 잘 구분해야 함!

# Axes
# Axes: 데이터 플로팅을 위한 영역, Figure에 포함되는 Artist
# Axis는 ticks와 tick labels을 제공하며, 이는 Axes의 데이터 범위에 대한 척도가 됨

# Axis(축)
# Axis: 플롯의 가로축이나 세로축을 나타내는 객체
# ticks(축의 표시)과 ticklabels(눈금 레이블을 지정하는 문자열)에 대한 규모와 제한을 설정
# 하나의 Axes 객체는 두 개 이상의 Axis 객체를 가짐
```

```python
# 여러가지 플롯을 하나의 Axes 객체에 표시할 때 

fig, ax0 = plt.subplots()
ax1 = ax0.twinx()      # twinx: x축을 공유하는 새로운 Axes 객체를 만드는 명령
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label = 'y0')
ax0.set_ylabel('y0')
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g:', label = 'y1')
ax1.set_ylabel('y1')
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```

```python
# Artist: Figure에 보이는 모든 것들

# Figure가 렌더링될 때, 모든 Artists가 그려짐
# 대부분의 Artists는 하나의 Axes에 묶여 있음
# 이러한 Artists는 여러 Axes에 공유되거나 이동할 수 없음
```
