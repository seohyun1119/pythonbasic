```python
# 파라미터의 기본 값이 True -> 결측치를 포함하지 않고 계산
# 인수로 skipna의 값을 False로 전달하면 결측값을 포함한 계산 결과 나옴
# numeric_only -> 수치형 컬럼(bool 포함)만 적용

titanic.sum(numeric_only=True)
```
